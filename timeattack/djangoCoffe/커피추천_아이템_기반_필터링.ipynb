```python
import pandas as pd
```

```python
category = pd.read_csv('/content/drinks_category.csv', encoding='cp949')
ratings = pd.read_csv('/content/new_drink_ratings.csv', encoding='cp949')

# pd.set_option('display.max_columns', 10)
# pd.set_option('display.width', 300)
```

```python
category
```

```python
ratings
```

```python
drink_ratings = pd.merge(ratings, category, on='drinkId')
drink_ratings
```

```python
title_user = drink_ratings.pivot_table('rating', index='userId', columns='title')
title_user = title_user.fillna(0)
title_user
```

```python
from sklearn.metrics.pairwise import cosine_similarity

user_based_collab = cosine_similarity(title_user, title_user)
print(user_based_collab)
```

```python
user_based_collab = pd.DataFrame(user_based_collab, index=title_user.index, columns=title_user.index)
```

```python
user_based_collab
```

```python
user_based_collab[1].sort_values(ascending=False)[:10]
```

```python
user = user_based_collab[2].sort_values(ascending=False)[:10].index[1]
user
```

```python
result = title_user.query(f'userId == {user}').sort_values(ascending=False, by=user, axis=1)
result
```

```python
top1 = result.
print(f'추천하는 음료는 {top1} 입니다!')
```
